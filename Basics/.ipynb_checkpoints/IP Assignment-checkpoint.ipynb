```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread("lamborghini.jpg", cv.IMREAD_GRAYSCALE)
img = cv.resize(img, (16,16), interpolation=cv.INTER_AREA)
```

```python
img.shape
```

```python
plt.imshow(img, cmap="gray")
plt.show()
```

```python
# Display the histogram of the original image
plt.hist(img.ravel(), 256, [0, 256])
plt.title('Original Image Histogram')
plt.show()
```

```python
# Apply histogram equalization
equalized_img = cv.equalizeHist(img)
```

```python
# Display the histogram of the equalized image
plt.hist(equalized_img.ravel(), 256, [0, 256])
plt.title('Equalized Image Histogram')
plt.show()
```

```python
# Plot original and equalized images Histogram side by side
plt.figure(figsize=(10, 4))

# Original image Histogram
plt.subplot(1, 2, 1)
plt.hist(img.ravel(), 256, [0, 256])
plt.title('Original Image Histogram')

# Equalized image Histogram
plt.subplot(1, 2, 2)
plt.hist(equalized_img.ravel(), 256, [0, 256])
plt.title('Equalized Image Histogram')
plt.show()
```

```python
# Plot original and equalized images side by side
plt.figure(figsize=(10, 4))

# Original image
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')

# Equalized image
plt.subplot(1, 2, 2)
plt.imshow(equalized_img, cmap='gray')
plt.title('Equalized Image')
plt.show()
```
